```python
from pandas_datareader import data
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
start = '2020-06-01'
end = '2022-05-25'

df = data.DataReader('^N225', 'yahoo', start, end)
```

```python
df.head(10)
```

```python
df.index
```

```python
data = df.index
price = df['Adj Close']
```

```python
plt.plot(data, price)
```

```python
plt.figure(figsize = (30, 10))
plt.plot(data, price)
```

```python
plt.figure(figsize = (30, 10))
plt.plot(data, price, label='Nikkei225')
plt.legend()
```

```python
plt.figure(figsize = (30, 10))
plt.plot(data, price, label='Nikkei225')
plt.title('N225', color = 'blue', backgroundcolor = 'white', size = 40, loc = 'center')
plt.legend()
```

```python
plt.figure(figsize = (30, 10))
plt.plot(data, price, label='Nikkei225')
plt.title('N225', color = 'blue', backgroundcolor = 'white', size = 40, loc = 'center')
plt.xlabel('date', color = 'black', size = 30)
plt.ylabel('price', color = 'black', size = 30)
plt.legend()
```

```python
span01 = 5
span02 = 25
span03 = 50

# rolling()  (window = )mean
df['sma1'] = price.rolling(window = span01).mean()
df['sma2'] = price.rolling(window = span02).mean()
df['sma3'] = price.rolling(window = span03).mean()
```

```python
pd.set_option('display.max_rows', None)  # 
df.head(500)
```

```python
plt.figure(figsize = (30, 10))
plt.plot(data, price, label = 'Nikkei225')
plt.plot(data, df['sma1'], label = 'sma1')
plt.plot(data, df['sma2'], label = 'sma2')
plt.plot(data, df['sma3'], label = 'sma3')

plt.title('N225(2020-2022)', color = 'green', backgroundcolor = 'white', size = 30, loc = 'center')
plt.xlabel('date', color = 'black', size = 30)
plt.ylabel('price', color = 'black', size = 30)
plt.legend()
```
